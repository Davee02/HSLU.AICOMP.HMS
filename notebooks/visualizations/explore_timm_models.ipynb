```python
import timm
```

```python
MODELS = [
    "tf_efficientnet_b0_ns",
    "tf_efficientnet_b1_ns",
    "tf_efficientnet_b2_ns",
    "tf_efficientnet_b3_ns",
    "resnet18",
    "resnet34",
    "resnet50",
    "resnet101",
    "inception_v3",
    "inception_v4",
]
```

```python
model = timm.create_model(
    MODELS[9],
    pretrained=True,
    in_chans=3,
    num_classes=6,
)
```

```python
model.default_cfg
```

```python
model_info = []

for model_name in MODELS:
    model = timm.create_model(
        model_name,
        pretrained=True,
        in_chans=3,
        num_classes=6,
    )

    num_params = sum(p.numel() for p in model.parameters())
    model_info.append(
        {
            "model_name": model_name,
            "num_parameters": num_params,
        }
    )
```
